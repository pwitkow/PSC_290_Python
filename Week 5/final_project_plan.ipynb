## EEG Power Analyses Tool

This is the plan for EEG power analyses tool. The goal of this class is to be able to conduct efficent power analyses and boot-strapping on small amount of preliminary EEG pilot data that can then tell the user how many subjects or trials per subject they will need to collect, if they want some specified level of power, assuming that the inherent noise in the data stays the same. Alternatively, it can tell them what power they will have, given n subject or n trials per subject, again if the noise in the data stays the same. 

The class should be able to take a directory in as an argument and find all of the eeg files specified by the user. It will organize this data into a main dataframe containing everything that needs to be acted on. Methods for visualiztion are primarily restricted to EEG components and boot straping analyses, but may included scalp potentials later. 

### Classes

#### *organize_data*
This is an optional class for organizing data into a usable data set. The goal is to be able to take a directory in as an argumnet and create a folder structure that the eeg_power class can use. It will pull inspiration from the BIDS format for fMRI, which organizes data into subject specific folders. 

##### Init Properties
1. directory **argument** : directory that need to be organized
2. file_tags **argument** : dictionary with keys 'subs', 'eeg_file', 'bhv_file'. This will tell the class how to index specific file types that need to be organized. If values are specified as None it will consider these files to be non-exsistent, or at least don't need to be orgnaized. 
3. out_dir **argument** : name of directory generated by calling the *run* method containing organized data. 
4. sub **argument** : List of specific subject ID's to be used if not all subjects are to be included in the new directory. 

##### Methods
1. *run* : this is the main method for class, which runs the data organization

#### *eeg_power*
The main class for conducting the power analyses.

##### Init properites
1. directory **arugment** : directory with subject data 
2. eeg_file **argument** : denotes the eeg file type that needs to be loaded in
2. eeg_tag **argument** : optional string argument that denotes a regex pattern for pulling EEG from mixed data set
3. data : pandas dataframe of loaded eeg data
4. boot_array : defaults to none. Modified by boot_strap method

##### Methods
These methods are ogranized into methods groups depending on what types of things they do. 

###### dataframe methods 
These methods contain nominal methods that are purley intended to act on the underlying dataframes so that 
users don't have to call the class data attribute directly
1. *to_array* : calls the dataframe property for array conversion
2. *subset* : used for subsetting dataframes by relvant values - takes arugments col, values
3. *add_col* : adds columns to the main dataset
4. *return_col* : returns selected columns 

###### data manipulation methods
These methods are for altering the data or generating new data based on the needs of the user
1. *detrend* : basic detrending of data overlaying scipy.signal.dertrend()
2. *average_potential* : returns mean potenial over time points - arguments for groupby and bin size
3. *boot_strap* : returns distribution of boot_strapped effects (not sure what this would be). Modifies the boot_array attribute to be an array 
of t-values
4. *power_analyses* : ags - nPeople, nTrials,effSize - returns power at specified levels. If input left as none or specified as array, the function will return an array that gives power for different level of that arugment, with specified arguments as constants.  

###### visualization methods
These methods are purley for visualizing and checking data
1. *erp_plot* : overlays seaborn or matplotlib time plot to show eeg components, with options for CI
2. *node_plot* : overlays a multi-axes plot for ERP x nodes
3. *boot_plot* : plots the boot_array attribute, indexing where true value falls
